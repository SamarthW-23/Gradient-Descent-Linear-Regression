```python
from src.data_loader import load_data
from src.linear_regression_gd import LinearRegressionGD

X, y = load_data()

model = LinearRegressionGD(learning_rate=0.01, n_iters=1000)
model.fit(X, y)

print("Learned weight:", model.weight)
print("Learned bias:", model.bias)
print("First 5 predictions:", model.predict(X[:5]))
print("First 5 losses:", model.loss_history[:5])
print("Last 5 losses:", model.loss_history[-5:])
```

```python
import matplotlib.pyplot as plt

# assuming X is shape (n_samples, 1)
X_flat = X[:, 0]

plt.scatter(X_flat, y, color="blue", label="Data")
y_line = model.predict(X)
plt.plot(X_flat, y_line, color="red", label="Fitted line")
plt.legend()
plt.show()
```

# Results

* The predicted weight for the regression line comes out to be 4.08763827, which is close to the actual weight.
* The predicted bias comes out to be 2.011754331638904, which is very close to the actual bias.

```python
plt.plot(model.loss_history)
plt.xlabel("Iteration")
plt.ylabel("MSE loss")
plt.title("Loss vs iterations")
plt.show()
```

# Loss VS Interation plot

The loss vs interation plot gives idea on how the loss decreases with every interation, and once it reaches a certain value, the graph flattens out and does not change after that.
